```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
from collections import Counter
```

```python
df=pd.read_csv("DataSet.csv")
df.head()
```

```python
a=Counter(df["Location"]).most_common(20)
city=[];count=[]
for i in range(len(a)):
    city.append(a[i][0])
    count.append(a[i][1])
```

```python
plt.figure(figsize = (10, 12))
plt.bar(city, count)
plt.xlabel("Locations")
plt.ylabel("Job Opportunities")
plt.xticks(city, rotation = "60")
plt.title("Top 20 JOB Opportunities")
plt.ylim(0,100 )
```

```python
a=Counter(df["Sector"]).most_common(20)
sector=[];count_sector=[]
for i in range(len(a)):
    sector.append(a[i][0])
    count_sector.append(a[i][1])
sector
```

```python
plt.figure(figsize = (10,20))
plt.bar(sector, count_sector)
plt.xlabel("Sector")
plt.ylabel("Job Opportunities")
plt.xticks(sector, rotation = "60")
plt.title("Job Opportunities in Top 20 Industries")
```

```python
#a=Counter(df["Salary"])
a=Counter(df["Job_Description"]).most_common(20)
JOB=[];count_sector=[]
for i in range(len(a)):
    JOB.append(a[i][0])
    count_sector.append(a[i][1])
JOB
```

```python
plt.figure(figsize = (10,20))
plt.bar(JOB, count_sector)
plt.xlabel("Profession")
plt.ylabel("Job Opportunities")
plt.xticks(JOB, rotation = "60")
plt.title("Job Opportunities in Top 20 Profession")
```
